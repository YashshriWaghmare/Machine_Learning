### convert the object type into numerical type

```python
from sklearn.preprocessing import LabelEncoder
def encoding(x_train, x_test):
    
    le = LabelEncoder()

    return (le.fit_transform(x_train), le.transform(x_test))
```

```python
df_train['workclass'], df_test['workclass'] = encoding(df_train['workclass'], df_test['workclass'])
df_train['education'], df_test['education'] = encoding(df_train['education'], df_test['education'])
df_train['maritalstatus'], df_test['maritalstatus'] = encoding(df_train['maritalstatus'], df_test['maritalstatus'])
df_train['occupation'], df_test['occupation'] = encoding(df_train['occupation'], df_test['occupation'])
df_train['relationship'], df_test['relationship'] = encoding(df_train['relationship'], df_test['relationship'])
df_train['race'], df_test['race'] = encoding(df_train['race'], df_test['race'])
df_train['sex'], df_test['sex'] = encoding(df_train['sex'], df_test['sex'])
df_train['native'], df_test['native'] = encoding(df_train['native'], df_test['native'])
df_train['Salary'], df_test['Salary'] = encoding(df_train['Salary'], df_test['Salary'])
```

```python
x_train = df_train.drop('Salary', axis = 1)
y_train = df_train['Salary']
```

```python
x_test = df_test.drop('Salary', axis = 1)
y_test = df_test['Salary']
```

### train the model

```python
def logistic_model():

    model = LogisticRegressionCV()
    model.fit(x_train, y_train)
    return model
```

```python
def naive_bayes():

    model = GaussianNB()
    model.fit(x_train, y_train)

    return model
```

```python
def knn():

    model = KNeighborsClassifier(n_neighbors=5)
    model.fit(x_train, y_train)

    return model
```

```python
def svm():

    model = SVC()
    model.fit(x_train, y_train)

    return model
```

```python
def decision_tree():

    model = DecisionTreeClassifier(max_depth=5)
    model.fit(x_train, y_train)

    return model
```

```python
def random_forest():

    model = RandomForestClassifier(n_estimators=100, max_depth=5)
    model.fit(x_train, y_train)

    return model
```

```python
def adaboost():

    model = AdaBoostClassifier()
    model.fit(x_train, y_train)

    return model
```

```python
models = [
    {"name": "Logistics Regression", "model": logistic_model()},
    {"name": "Naive Bayes", "model": naive_bayes()},
    {"name": "KNN", "model": knn()},
    {"name": "SVM", "model": svm()},
    {"name": "Decision Tree", "model": decision_tree()},
    {"name": "Random Forest", "model": random_forest()},
    {"name": "AdaBoost", "model": adaboost()},
]
```

### evaluate the model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate(model, name, x, y):
    y_pred = model.predict(x)

    accuracy = accuracy_score(y,y_pred)
    precision = precision_score(y,y_pred)
    recall = recall_score(y,y_pred)
    f1 = f1_score(y,y_pred)

    return name , accuracy, precision, recall, f1
```

```python
metrics_train = []

for model_info in models:

    metrics_train.append(evaluate(model_info['model'], model_info['name'], x_train, y_train))

metrics_train_df = pd.DataFrame(metrics_train, columns=['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
metrics_train_df
```

```python
metrics_test = []

for model_info in models:

    metrics_test.append(evaluate(model_info['model'], model_info['name'], x_test, y_test))

metrics_test_df = pd.DataFrame(metrics_test,  columns=['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
metrics_test_df
```
