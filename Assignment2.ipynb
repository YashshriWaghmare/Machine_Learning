### Que 1.
The delivery_time.csv dataset contains information about Delivery time and Sorting Time. We have to prepare linear regression model to predict Sorting time on the basis of Delivery Time. Also evaluate the model by using evaluation metrics. Do proper EDA & visualize the result.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./delivery_time.csv')
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.corr()
```

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
x=df.drop('Sorting Time', axis=1)
y=df['Sorting Time']
```

```python
plt.scatter(x['Delivery Time'],y, label="Sorting Time")
plt.xlabel("Delivery Time")
plt.ylabel("Sorting Time")
plt.title("Delivery Time vs Sorting Time")
plt.legend()
plt.grid()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred_train = model.predict(x_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

# calcuate the r2 score
r2 = r2_score(y_train, y_pred_train)
print(f"r2 score: {r2:.2f}")
```

```python
from sklearn.metrics import mean_absolute_error

# calculate the mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print(f"mean absolute error: {mae:.2f}")
```

```python
from sklearn.metrics import mean_absolute_percentage_error

# calculate the mape
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"mean absolute percentage error = {mape * 100:.2f}%")
```

```python
from sklearn.metrics import mean_squared_error

# calculate the mse
mse = mean_squared_error(y_test, y_pred)
print(f"mean squared error: {mse:.2f}")
```

```python
from sklearn.metrics import root_mean_squared_error

# calculate the rmse
rmse = root_mean_squared_error(y_test, y_pred)
print(f"root mean squared error: {rmse:.2f}")
```

```python
from sklearn.metrics import r2_score

# calcuate the r2 score
r2 = r2_score(y_test, y_pred)
print(f"r2 score: {r2:.2f}")
```

```python
plt.scatter(x_train,y_train,label="Sorting Time")
plt.scatter(x_train,y_pred_train,color="red")
plt.plot(x_train,y_pred_train,color="red",label="Regression line")
plt.xlabel("Delivery Time")
plt.ylabel("Sorting Time")
plt.legend()
plt.grid()
```

```python
plt.scatter(x_test,y_test,label="Sorting Time")
plt.scatter(x_test,y_pred,color="red")
plt.plot(x_test,y_pred,color="red",label="Regression line")
plt.xlabel("Delivery Time")
plt.ylabel("Sorting Time")
plt.legend()
plt.grid()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./SAT-GPA.csv')
```

```python
plt.scatter(x,y)
```
