### Que.1
The iris.csv dataset contains measurements of iris flowers belonging to three different species:
Each flower is described using four numerical features:
- Sepal Length (cm)
- Sepal Width (cm)
- Petal Length (cm)
- Petal Width (cm)

The dataset consists of 150 records & you have to predict the species of iris flowers.
Prepare model by using Logistic Regression, KNN, SVM, Decision Tree, Random Forest, AdaBoost, XGBoost & Calculate Accuracy by confusion matrix, Calculate accuracy score, precision score, recall score, f1 score.
Also Prepare summary of all models & Select best of them.


### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
```

### load the data

```python
df = pd.read_csv('iris.csv')
df.head(2)
```

### exploratory data analysis

```python
df.info()
```

```python
df.describe()
```

```python
# convert str to numerical type

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['species'] = le.fit_transform(df['species'])
```

```python
df.head(2)
df.info()
```

```python
sns.heatmap(df.corr(), annot = True)
```

```python
df.corr()
```

```python
df['species'].value_counts()
```

### split the data

```python
# split the data into x and y
x = df.drop('species', axis = 1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x , y, train_size= 0.80, random_state= 42)
```

### create the model

```python
def logistic_model():

    model = LogisticRegressionCV()
    model.fit(x_train, y_train)

    return model
```

```python
def knn_model():

    model = KNeighborsClassifier()
    model.fit(x_train, y_train)

    return model
```

```python
def svm_model():

    model = SVC()
    model.fit(x_train, y_train)

    return model
```

```python
def decision_tree_model():

    model = DecisionTreeClassifier()
    model.fit(x_train, y_train)

    return  model
```

```python
def random_forest_model():

    model = RandomForestClassifier()
    model.fit(x_train, y_train)

    return model
```

```python
def adaboost_model():

    model = AdaBoostClassifier()
    model.fit(x_train, y_train)

    return model
```

```python
def xgboost_model():

    model = XGBClassifier()
    model.fit(x_train, y_train)

    return model
```

```python
models = [
    {"name": "Logistic Regression", "model" : logistic_model()},
    {"name": "KNN" , "model": knn_model()},
    {"name": "SVM", "model": svm_model()},
    {"name": "Decision Tree", "model": decision_tree_model()},
    {"name": "Random Forest ", "model": random_forest_model()},
    {"name": "AdaBoost", "model": adaboost_model()},
    {"name": "XGB", "model": xgboost_model()}
]
```

### evaluate the model

```python
from sklearn.metrics import confusion_matrix
model = logistic_model()
y_pre = model.predict(x_test)
cm = confusion_matrix(y_test, y_pre)
print(cm)
```

```python
from sklearn.metrics import  accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model, name, x, y):

    y_pred = model.predict(x)

    accuracy = accuracy_score(y, y_pred)
    precision = precision_score(y,y_pred, average = 'weighted')
    recall = recall_score(y, y_pred, average = 'weighted')
    f1 = f1_score(y, y_pred, average = 'weighted')

    return name, accuracy, precision, recall, f1
```

### Training Dataset

```python
metrics_train = []

for model_info in models:

    metrics_train.append(evaluate_model(model_info['model'], model_info['name'], x_train, y_train))

metrics_train_df = pd.DataFrame(metrics_train, columns=['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
metrics_train_df
```

### Testing Dataset

```python
metrics_test = []

for model_info in models:

    metrics_test.append(evaluate_model(model_info['model'], model_info['name'], x_test, y_test))

metrics_test_df = pd.DataFrame(metrics_test, columns=['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
metrics_test_df
```
